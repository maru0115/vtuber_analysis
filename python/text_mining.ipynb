```python
list_words    = []
list_id       = []
list_df_title = []
list_df_type  = []

for i, title in enumerate(list_title):

   list_nouns     = [line for line in mecab.parse(title).splitlines() if "名詞" in line.split()[-1]]
   list_add_words = [nouns.split()[0] for nouns in list_nouns]

   list_words    += [nouns.split()[0] for nouns in list_nouns]
   list_id       += [i]            * len(list_add_words)
   list_df_type  += [list_type[i]] * len(list_add_words)
   list_df_title += [title]        * len(list_add_words)
```

```python
df = pd.DataFrame({
    "word"  : list_words,
    "id"    : list_id,
    "type"  : list_df_type,
    "title" : list_df_title
})

df_count = df.query(f"word not in {ng_list}").groupby("word").count().filter(["id"]).rename(columns = {"id":"count"})
```

```python
df.merge(df_count, how="inner", on="word").reindex(
    columns=["word","count","id","title"]
).sort_values(
    ["count","word","id"], ascending=[False,True,True]
).to_csv("./words.csv", index=False)
```
